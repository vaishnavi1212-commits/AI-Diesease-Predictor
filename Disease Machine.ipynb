```python
import matplotlib.pyplot as plt

time = ["8am", "10am", "12pm", "2pm", "4pm", "6pm"]
heart_rate = [88, 92, 110, 105, 98, 90]
blood_sugar = [140, 155, 190, 180, 160, 150]

plt.figure(figsize=(8, 5))
plt.plot(time, heart_rate, marker='o', label='Heart Rate (bpm)', color='red')
plt.plot(time, blood_sugar, marker='s', label='Blood Sugar (mg/dL)', color='blue')
plt.title("Simulated Health Data Throughout the Day")
plt.xlabel("Time")
plt.ylabel("Measurement")
plt.legend()
plt.grid(True)

plt.savefig("simulated_health_chart.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

time = ["8am", "10am", "12pm", "2pm", "4pm", "6pm"]
x = np.arange(len(time))

heart_rate = [78, 85, 95, 102, 98, 92]
blood_sugar = [120, 140, 185, 160, 145, 135]
bp_systolic = [130, 138, 142, 145, 140, 135]

width = 0.25

plt.figure(figsize=(10, 6))
plt.bar(x - width, heart_rate, width=width, label='Heart Rate', color='red')
plt.bar(x, blood_sugar, width=width, label='Blood Sugar', color='blue')
plt.bar(x + width, bp_systolic, width=width, label='BP (Systolic)', color='green')

plt.xticks(x, time)
plt.title("Patient Health Vitals - Bar Chart")
plt.xlabel("Time of Day")
plt.ylabel("Measurement")
plt.legend()
plt.grid(True)

plt.savefig("bar_chart_health_vitals.png")
plt.show()
```
